```python

import pandas as pd

file_path = '../shared_data_read_only/Data/AfriKids/InternalData/1. Master Stats 2023.xlsx'

df = pd.read_excel(file_path, engine='openpyxl')

df

import matplotlib.pyplot as plt

regions = df.columns[1:]  
population_data = df.loc[df[0] == 'Population']

plt.figure(figsize=(10, 6))
plt.bar(regions, population_data.iloc[0, 1:])
plt.title('Population Distribution by Region')
plt.xticks(rotation=45)
plt.show()


```
