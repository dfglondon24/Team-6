```python
district_list = age_sex.iloc[:, 0].unique().tolist()
district_list = [district_list[2]] + district_list[101:]
gender_list = ['Both sexes', 'Male', 'Female']
age_list = ['Total', '< 1'] + [i for i in range(1, 95)] + ['95+']
locality_list = ['Total', 'Urban', 'Rural']
number_list = ['Number', 'Percent']

mask1 = ~age_sex.iloc[:, 0].isin(district_list)
mask2 = ~age_sex.iloc[:, 1].isin(['Total'])
mask3 = ~age_sex.iloc[:, 1].isin(gender_list)
mask4 = ~age_sex.iloc[:, 1].isin(number_list)
mask5 = ~age_sex.iloc[:, 0].isna()
# Combine the masks using the bitwise AND operator
combined_mask = mask1 & mask2 & mask3 & mask4 & mask5

# Use the combined mask to filter the DataFrame
age_sex_filtered = age_sex[combined_mask]

age_sex_filtered = age_sex_filtered.dropna(axis=1, how='all').drop(age_sex_filtered.columns[-1], axis = 1).reset_index(drop = True)

# Display the DataFrame after removing columns with all NaN values
age_sex_filtered


multi_index = pd.MultiIndex.from_product(
    [number_list, locality_list, gender_list, district_list,  age_list],
    names=['Number/Percent', 'Locality', 'Gender', 'District', 'Age']
)

# Create an empty DataFrame with the multi-index
df_age = pd.DataFrame(index=multi_index, columns = ['Population'])


column = 0
index = 0
counter = 0
for i in range(474912):
    if age_sex_filtered.iloc[index, 0] == '95+':
        df_age.iloc[i] = age_sex_filtered.iloc[index, column + 1]
        column+= 1
        index -= counter
        counter = 0
        if column + 1 >= age_sex_filtered.shape[1]:
            column = 0
            index += 98
    else:
        df_age.iloc[i] = age_sex_filtered.iloc[index, column + 1]
        counter += 1
        index += 1
```
